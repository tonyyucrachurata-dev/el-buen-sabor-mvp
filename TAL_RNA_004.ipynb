```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder, LabelBinarizer
```

```python
postulantes= pd.DataFrame({
    'lenguaje':['java','js','php','java','java','php','js','java'],
    'perfil':['back','front','back','back','back','back','front','back'],
    'exp':[5, 1, 3, 1, 3, 5, 5, 2],
    'sueldo':[5000, 1000,3000,1000,3000,5000,5000,2000],
    'estado':['apto','no apto','no apto','no apto','apto','no apto','no apto','no apto']
})
postulantes
```

```python
x_cat = postulantes[['lenguaje','perfil']]
x_cat
```

```python
x_num = postulantes[['exp','sueldo']]
x_num
```

```python
y_bin = postulantes[['estado']]
y_bin
```

```python
cat_encoder = OneHotEncoder(sparse_output= False) #PARA VARIABLES CATEGÓRICAS NOMINALES
num_scaler = StandardScaler() #PARA VARIABLES NUMÉRICAS
y_binarizer = LabelBinarizer() #PARA VARIABLES CATEGÓRICAS ORDINALES
```

```python
x_cat_encoder = cat_encoder.fit_transform(x_cat)
x_cat_encoder
```

```python
x_num_scaler = num_scaler.fit_transform(x_num)
x_num_scaler
```

```python
y_binarizer = y_binarizer.fit_transform(y_bin)
y_binarizer
```

```python
postulantes_normalizer = np.hstack((x_cat_encoder,x_num_scaler,y_binarizer))
postulantes_normalizer
```

```python
planilla_pandas = pd.DataFrame(postulantes_normalizer)
planilla_pandas
```

```python
x = postulantes_normalizer[:, 0:7]
x
```

```python
y = postulantes_normalizer[:, -1]
y
```

```python
print(x.shape)
print(y.shape)
```

```python
import tensorflow as tf
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(20, activation='relu', input_shape=(7,)),
    tf.keras.layers.Dense(10, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(x, y, epochs=50)
```

```python
y_pred = model.predict(x)
print('DAtos Reales',y) #Datos Reales
print('Predicciones',np.round(y_pred)) #Predicciones
```

```python
from sklearn.metrics import accuracy_score, classification_report
print(accuracy_score(y, np.round(y_pred)))
print(classification_report(y, np.round(y_pred)))
```
